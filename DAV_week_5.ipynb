```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data Simulation
np.random.seed(42)

bluedart_times = np.random.normal(24, 4, 1000)
delhivery_times = np.random.normal(28, 6, 1000)
dtdc_times = np.random.normal(26, 5, 1000)

delivery_times = np.concatenate([bluedart_times, delhivery_times, dtdc_times])
courier_partners = (['BlueDart'] * 1000 + ['Delhivery'] * 1000 + ['DTDC'] * 1000)

delivery_df = pd.DataFrame({
    'Delivery Time (hours)': delivery_times,
    'Courier Partner': courier_partners
})

# Plot Generation and Aesthetic Customization
plt.figure(figsize=(10, 6))
sns.boxplot(x='Courier Partner', y='Delivery Time (hours)', data=delivery_df, palette='viridis', hue='Courier Partner')
plt.ylabel('Delivery Time (Hours)')
plt.title('Distribution of Delivery Times by Courier Partner')
plt.show()
```

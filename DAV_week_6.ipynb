```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(42)
sample_weights = np.random.lognormal(mean=3.5, sigma=0.3, size=100)

weights_series = pd.Series(sample_weights)

plt.figure(figsize=(10, 6))
plt.hist(weights_series, bins=15, edgecolor='black')
plt.title('Distribution of Gym Member Weights (Sample Data)')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

mean_weight = weights_series.mean()
median_weight = weights_series.median()

print(f"Mean Weight: {mean_weight:.2f} kg")
print(f"Median Weight: {median_weight:.2f} kg")

if mean_weight > median_weight:
    print("The histogram is likely right-skewed (mean > median).")
elif mean_weight < median_weight:
    print("The histogram is likely left-skewed (mean < median).")
else:
    print("The histogram is likely symmetrical (mean = median).")
```
